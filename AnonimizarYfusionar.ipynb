```python
import pandas as pd
import os
import re
from glob import glob
from tqdm import tqdm

SRC_DENUNCIOS="./src/RelatoDenuncio/"
```

```python
df_siniestros       = pd.read_csv("DS_SINIESTROS.CSV", sep=";", encoding="UTF-8")
df_poliza_cobertura = pd.read_csv("DS_POLIZA_COBERTURA.CSV", sep=";", encoding="UTF-8")
df_poliza_vehiculo  = pd.read_csv("DS_POL_VEHICULO.CSV", sep=";", encoding="UTF-8")

#fichero nuevo hecho a mano para estandarizar
df_cobertura_h = pd.read_csv("COBERTURAS_HOMOLOGADAS.CSV", sep=";", encoding="cp1252")

homologacion = {}
for i,j in zip(df_cobertura_h['COBERTURA'],df_cobertura_h['NOMBRE_ESTANDARIZADO']):
    homologacion[i] = j

#estandarizando nombres de coberturas.
df_siniestros['COBERTURA_HOMOLOGA'] = [ homologacion[i] for i in df_siniestros['COBERTURA_AFECTADA']]
df_poliza_cobertura['COBERTURA_HOMOLOGA'] = [ homologacion[i] for i in df_poliza_cobertura['COBERTURA'] ]
```

```python
def getRelato(id_siniestro):
    path_relato = SRC_DENUNCIOS + "{}.txt".format(id_siniestro)
    fichero_relato = open(path_relato,"r",encoding="utf-8")
    lineas = fichero_relato.readlines()
    #elimina blancos execivos
    lineas = " ".join([re.sub( "\ +"," ", l ).strip() for l in lineas]).strip()
    #elimina nombre de la empresa
    lineas = re.sub(" BCI| BCI SEGUROS","  ", lineas)
    #elimina presencia de numeros que pudiesten ser ruts
    lineas = re.sub("\d\d\d\d\d\[d|\d\d]*[\-|\s]*", "", lineas)
    return lineas

def getCobSin(id_siniestro):
    return list(df_siniestros[ df_siniestros['ID_SINIESTRO']==id_siniestro ]['COBERTURA_HOMOLOGA'].unique())

def getGastos(id_siniestro):
    provisiones = []
    gastos_reales = []
    
    for cob in getCobSin(id_siniestro):
        provision_x= sum(df_siniestros[(df_siniestros['ID_SINIESTRO']==id_siniestro)&(df_siniestros['COBERTURA_HOMOLOGA']==cob) ]['PROVISION_INICIAL'])
        gasto_x= sum(df_siniestros[(df_siniestros['ID_SINIESTRO']==id_siniestro)&(df_siniestros['COBERTURA_HOMOLOGA']==cob) ]['GASTO_REAL'])
        
        provisiones.append(provision_x)
        gastos_reales.append(gasto_x)

    return provisiones, gastos_reales     

def getPoliza(id_siniestro):
    return df_siniestros[ df_siniestros['ID_SINIESTRO']==id_siniestro ]['ID_POLIZA'].iloc[-1]

def getCobPol(id_poliza):
    return list(df_poliza_cobertura[df_poliza_cobertura['ID_POLIZA']==id_poliza]['COBERTURA_HOMOLOGA'].unique())

def getVehiculo(id_poliza):
    registro = df_poliza_vehiculo[ df_poliza_vehiculo['ID_POLIZA']==id_poliza ]
    if len(registro)>0:
        return registro.iat[0,2], registro.iat[0,3], str(int(registro.iat[0,1]))
    else:
        return("sin identificar","sin identificar", "sin identificar")
```
